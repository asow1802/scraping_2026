```python
#Importation des données
import pandas as pd

achat = pd.read_csv("Achat.csv")
loc = pd.read_csv("LocationParMois.csv")
prix_loc = pd.read_csv("PrixLocationSurDemande.csv")
prix_dem = pd.read_csv("PrixSurDemande.csv")

# Nombre de références uniques
print("Achat :", achat["Réf. client :"].nunique())
print("Location :", loc["Réf. client :"].nunique())
print("Prix location sur demande :", prix_loc["Réf. client :"].nunique())
print("Prix sur demande :", prix_dem["Réf. client :"].nunique())

# Vérification que toutes les références sont dans la base principale
set_achat = set(achat["Réf. client :"])
set_loc = set(loc["Réf. client :"])
set_prix_dem = set(prix_dem["Réf. client :"])
set_prix_loc = set(prix_loc["Réf. client :"])

print("Achat ⊂ PrixLocationSurDemande :", set_achat.issubset(set_prix_loc))
print("Location ⊂ PrixLocationSurDemande :", set_loc.issubset(set_prix_loc))
print("PrixSurDemande ⊂ PrixLocationSurDemande :", set_prix_dem.issubset(set_prix_loc))
```

```python

import numpy as np
# Copie de la base pour la nettoyer 
achat_clean = achat.copy()

# Nettoyage du kilométrage 
def clean_km(x):
    if isinstance(x, str):
        return int(x.replace(" km", "").replace(" ", "").replace(",", ""))
    return np.nan

achat_clean["Kilométrage"] = achat_clean["Kilométrage"].apply(clean_km)

# Nettoyage des dates
achat_clean["Date de 1ere immatriculation"] = pd.to_datetime(
    achat_clean["Date de 1ere immatriculation"], errors="coerce", dayfirst=True
)

# Nettoyage du prix 
achat_clean["Prix"] = pd.to_numeric(achat_clean["Prix"], errors="coerce")

# Nettoyage du poids 
def clean_weight(x):
    if isinstance(x, str):
        return float(x.replace(" Tonnes", "").replace(",", "."))
    return np.nan

achat_clean["Poids à vide"] = achat_clean["Poids à vide"].apply(clean_weight)
achat_clean["PTC"] = achat_clean["PTC"].apply(clean_weight)

# Nettoyage des colonnes texte 
colonnes_texte = ["Description", "Energie", "Marque", "Modèle", "Pays", 
                  "Vendeur", "Localisation", "Etat"]

for col in colonnes_texte:
    achat_clean[col] = achat_clean[col].astype(str).str.strip()

# Suppression des doublons 
achat_clean.drop_duplicates(subset=["Réf. client :"], inplace=True)

# Suppression de la colonne Photo
if "Photo" in achat_clean.columns:
    achat_clean.drop(columns=["Photo"], inplace=True)


# Export de la table propre
achat_clean.to_csv("Achat_clean.csv", index=False)

achat_clean.head()
```
