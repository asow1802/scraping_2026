```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import time
import random
import numpy as np

from selenium import webdriver
from selenium.webdriver.common.by import By

# Récupération du contenu de la page d'accueil du site de vente
url = "https://www.mascus.com/+/catalogs=cargo-transport&categories=trucks&continentcodes=150/1,relevance,search.html"

# Initialiser le driver
driver = webdriver.Chrome()
driver.get(url)
time.sleep(random.uniform(2, 5))  # Attendre un temps aléatoire

# Récupérer tous les href de pages de produits
hrefs = driver.find_elements(By.XPATH, "//a[@class='product-link']")
hrefs = [href.get_attribute("href") for href in hrefs]
print(hrefs)
```
